```python
from youtube_transcript_api import (
    YouTubeTranscriptApi,
    NoTranscriptFound,
    TranscriptsDisabled,
    VideoUnavailable
)

from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_huggingface import HuggingFaceEmbeddings

from langchain_community.vectorstores import FAISS

from langchain_core.prompts import PromptTemplate
```

STEP 1 INDEXING


```python

def ingest_video_transcript(video_id):
    try:
        api = YouTubeTranscriptApi()

        transcript = api.fetch(video_id)

        lines = []

        for chunk in transcript:
            lines.append(chunk.text)

        return " ".join(lines)

    except Exception as e:
        return f"Error: {e}"
```

```python
video_id="wjZofJX0v4M"
transcript_text = ingest_video_transcript(video_id)
print(transcript_text)  # Print the first 500 characters of the transcript
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = text_splitter.create_documents([transcript_text])
print(f"Number of chunks created: {len(chunks)}")
```

```python
chunks
```

```python
# embdeddings generation
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
#vector store
vector_store=FAISS.from_documents(chunks,embeddings)
```

```python
vector_store.index_to_docstore_id
```

```python
# retrivel
retrieved_document = vector_store.docstore.search('1aff2c17-11b4-439e-9b88-57f4afc1248a')
print(retrieved_document)
```

```python
retriver=vector_store.as_retriever(search_type='similarity', search_kwargs={'k':5})
```

```python
retriver.invoke('what are transformer')
```

STEP 3 Augmentation


```python
prompt=PromptTemplate(
    template="""you are helpful assistant .
    answer only from the provided transcripts context.
    if the context is insufficient just say i don't know
    Rules:
Rules:
- Return only the final answer.
- Do not return the context.
- give answer in bullet point



    {context}
    Question: {question}""",
    input_variables=['context','question']
)
```

```python
question=" is something discussed related to transformer? if yes then what?"
retrieved_docs=retriver.invoke(question)
```

```python
retrieved_docs
```

```python
context_text="/n/n".join(doc.page_content for doc in retrieved_docs)
```

```python
context_text
```

```python
final_prompt=prompt.format(context=context_text,question=question)
```
